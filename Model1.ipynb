```python
import pandas as pd
import seaborn as sns 
import matplotlib.pyplot as plt
import numpy as np
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score
np.random.seed(42)
```

```python
df=pd.read_csv("D:/train/Train.csv") 
test = pd.read_csv("D:/train/Test.csv")
subm= pd.read_csv("D:/train/SampleSubmission.csv")

df.head(10)
```

```python
df['Unit Price'] = pd.to_numeric(df['Unit Price'], errors='coerce')
df['Quantity'] = pd.to_numeric(df['Quantity'], errors='coerce')
df['Delivery Charge'] = pd.to_numeric(df['Delivery Charge'], errors='coerce')
test['Delivery Charge']= pd.to_numeric(test['Delivery Charge'],errors='coerce')
```

```python
cash = df['Payment Method']=='CASH_PAY' 
df['Cash'] = cash
test['Cash']= cash
df.head(10)
```

```python
card = df['Payment Method']=='CARD_PAY' 
df['Card'] = card
test['Card']= card
df.head(10)
```

```python
df['Cash'] = (df['Cash'] == True ).astype(int)
df['Card'] = (df['Card'] == True ).astype(int)
df.head(10)
```

```python
del df["Total Price"] 
del test["Total Price"] 
```

```python
TP = df["Unit Price"]*df["Quantity"] 
df["Total Price"]= TP 
test["Total Price"]= TP
```

```python
df.head(20)
```

```python
new_df = df.groupby('Customer ID').agg({
    'Order ID': 'count',
    'Quantity': 'sum',
    'Merchant': 'count',
    'Total Price': 'sum',
    'Purchase Again' : 'max',
    'Delivery Charge':'sum',
     'Cash' : 'sum',
     'Card' : 'sum',
    
    

})

nw_test = test.groupby('Customer ID').agg({
    'Order ID': 'count',
    'Quantity': 'sum',
    'Merchant': 'count',
    'Total Price': 'sum', 
    'Delivery Charge':'sum',
    'Cash' : 'sum',
     'Card' : 'sum',
    
    
    }) 
```

```python
nw_test.info()
```

```python

new_df.head(10)
```

```python
X= new_df.drop('Purchase Again',axis=1)
y=new_df['Purchase Again'] 
```

```python
model= RandomForestClassifier(random_state=42) 
model.fit(X,y)
```

```python
y_hat = model.predict(nw_test) 
subm['Target']= y_hat 
subm.to_csv("D:/train/Data2.csv")
subm.head(10)
```
